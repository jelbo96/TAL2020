### Modelos N-Grama

#### Obtener un datos de CNN

```python
import pandas as pd

DATASET_CSV="datasets/CNNCHILE_RAW.csv"

df = pd.read_csv(DATASET_CSV,error_bad_lines=False)
df[:5]
```

```python
#df_short = df['body'][16400:]

df_test = df['body']

array_cnn = []
```

```python
for news in df_test:
    ## Noticia completa
    if(type(news)!= float):  #Hay noticias que son solo un numero
    
        array_cnn.append(news.split())
    
```

```python
array_cnn[500]
```

```python
#from nltk.corpus import reuters
from nltk import bigrams, trigrams
from collections import Counter, defaultdict

# Create a placeholder for model
model = defaultdict(lambda: defaultdict(lambda: 0))

# Count frequency of co-occurance  
for sentence in array_cnn:
    for w1, w2, w3 in trigrams(sentence, pad_right=True, pad_left=True):
        model[(w1, w2)][w3] += 1
 
# Let's transform the counts to probabilities
for w1_w2 in model:
    total_count = float(sum(model[w1_w2].values()))
    for w3 in model[w1_w2]:
        model[w1_w2][w3] /= total_count
```

### Modelos Word-embeddings

```python
from gensim.models import word2vec
import pandas as pd
```

#### Probar con dataset reino unido 

```python
DATASET="datasets/itv-reinounido.csv" #ej: nprnews-eeuu.csv, radionewzealand.csv
df = pd.read_csv(DATASET,delimiter="|")
df.head(5)
```

```python
df_10000=df.head(1000)
```

```python
# preprocesar la columna "text" para tener un dataset de entrenamiento (lista de tokens)
import spacy

nlp = spacy.load('en_core_web_sm')
spacy_stopwords = spacy.lang.en.stop_words.STOP_WORDS
```

```python
train_dataset=""

for index,row in df_10000.iterrows():
    # Text of the news
    text=row[1]
    
    #preprocesamiento spacy
    doc = nlp(text)
    
    for token in doc:
        #print(str(token).lower()+str(token.pos_))
        if (str(token.pos_)!="SPACE" and str(token.pos_)!="PUNCT"):
            train_dataset=train_dataset+str(token).lower()+" "

    #break
```

```python
import codecs

file = codecs.open("train_dataset_united_kingdom.txt", "w", "utf-8")
file.write(train_dataset)
file.close()
```

```python
with open('train_dataset_united_kingdom.txt', 'w') as file: # Use file to refer to the file object
    file.write(train_dataset)
```

```python
#training word2vec

tokens = word2vec.Text8Corpus('train_dataset_united_kingdom.txt')

model = word2vec.Word2Vec(tokens,size=200,hs=1)
```

```python
print(model)
```

```python
model.wv.most_similar(positive=['woman'],topn=5)
```

```python
model.wv.most_similar(positive=['man'],topn=5)
```

```python
model.save("itv_reinounido_wordembedding")
```
